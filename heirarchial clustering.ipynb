```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('D:\\udemy\\A-Z data sets\\P14-Part4-Clustering\\Section 26 - Hierarchical Clustering\\Python\\Mall_Customers.csv')
```

```python
x=df.iloc[:,[3,4]]
```

```python
### we use scipy library which has the hirarchial clustering tool
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))  ## ward method minimizes the variance in each of the clusters
```

### the above step shows us the number of clusters we need. that is abvious 5

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean')
y_pred=hc.fit_predict(x)
```

```python
plt.scatter(x.iloc[y_pred==0,0],x.iloc[y_pred==0,1],s=100,c='red',label='c1')
plt.scatter(x.iloc[y_pred==1,0],x.iloc[y_pred==1,1],s=100,c='blue',label='c2')
plt.scatter(x.iloc[y_pred==2,0],x.iloc[y_pred==2,1],s=100,c='green',label='c3')
plt.scatter(x.iloc[y_pred==3,0],x.iloc[y_pred==3,1],s=100,c='cyan',label='c4')
plt.scatter(x.iloc[y_pred==4,0],x.iloc[y_pred==4,1],s=100,c='magenta',label='c5')
plt.xlabel('fetures')
plt.ylabel('% probability')
```
